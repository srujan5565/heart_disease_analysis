```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Assuming X and y are your feature and target data

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=3)

# Create and train the Logistic Regression model with a larger max_iter
model = LogisticRegression(max_iter=1000)  # Set max_iter to a larger value

# Ensure that X_train and X_test have valid feature names associated with their columns
# For training data
x_train = pd.DataFrame(x_train, columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal'])
y_train = pd.Series(y_train)

# For testing data
x_test = pd.DataFrame(x_test, columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal'])
y_test = pd.Series(y_test)

# Train the Logistic Regression model using pandas DataFrames
model.fit(x_train, y_train)

# Perform predictions on the test data
y_pred = model.predict(x_test)
```
